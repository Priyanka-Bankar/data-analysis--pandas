**Q1. What is the average median income of the data set and check the distribution of data using appropriate plots. Please explain the distribution of the plot.**

- It is a univariate analyis.
- column datatype is numerical(continuous).
- we can use box plot or histogram to see distrinution.
- using box plot we can also see skewness of the data.

```python
# average median income = 3.870671
house["median_income"].mean()
```

```python
# plot using seaborn
sns.histplot(data=house,x='median_income')
```

```python
# plot using matplotlib.
plt.hist(house["median_income"], color='blue',edgecolor='black')
plt.title("Average Median Income")
plt.xlabel("Median Income")
plt.ylabel("Frequencies")
plt.grid(False)
plt.show()
```

```python
# using boxplot
sns.boxplot(data=house,y='median_income')
```

## Interpretation:

- From above histogram or box plot we can see that average median_income is 3.78 and plot is positively skewed/right skewed.
- which means median(3.53) is less than mean(3.87) 
- from this we can interpreat that their are some people which are receiving higher salries/income.
- outliers also present .

**Q10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.**
